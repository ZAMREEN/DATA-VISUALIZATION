```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
x=df[["CustomerAge","CustomerGender","PurchaseFrequency","CustomerSatisfaction"]]
y=df["PurchaseIntent" ]
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test) 
```

```python
svm_model=SVC(kernel="linear",random_state=42)
svm_model.fit(x_train,y_train)
```

```python
y_pred=svm_model.predict(x_test)
```

```python
accuracy=accuracy_score(y_pred,y_test)
accuracy
```

```python
report=classification_report(y_test,y_pred)
print(f'classification Report:\n{report}')
```

```python
conf_matrix=confusion_matrix(y_test,y_pred)
print(f'confusion matrix:\n{conf_matrix}')
```

```python
new_data = np.array([[18,0,2,1]])  # New input
new_data_scaled = scaler.transform(new_data)  # Apply scaling

prediction = svm_model.predict(new_data_scaled)
prediction
```
